## Create Panda Data Frame from the Dictionary

```python
boston = pd.DataFrame(data=boston_housing_raw['data'],columns=boston_housing_raw['feature_names'])
```

```python
boston['PRICE'] = boston_housing_raw['target']
```

We can see that the highest correlation that influence the price is total number of room and the highest factor that could decrease the housing price is the % lower status of the population.
By implenting **linear regression**, we can predict how big is the influence of each data to the estimated price

## Implementing Linear Regression

```python
# importing train test split variable
from sklearn.model_selection import train_test_split
```

```python
# creating x and y variable
x = boston.drop('PRICE',1)
y = boston['PRICE']
```

```python
# check x and y data
x.head()
```

```python
y
```

**Build training set and test set**

```python
#creating train and test data set by spliting on ratio 37% test set and 63% train set
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
x_train.info()
```

```python
# import linear reg object
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
# train the data
lm.fit(x_train, y_train)
```

```python
lm.intercept_
```

## Predicting Test Data

```python
predict = lm.predict(x_test)
```

```python
y_test.head()
```

```python
g = sns.scatterplot(y_test, predict)
g.set_xlabel('Y test')
g.set_ylabel('Predicted Y')
```

## Residuals

```python
sns.distplot((y_test - predict), bins=40)
```

## Calculate the Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predict))
print('MSE:', metrics.mean_squared_error(y_test, predict))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predict)))
```
